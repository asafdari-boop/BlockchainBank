#Trapezoidal rule

$I(a, b)=h[\frac{1}{2} f(a) + \frac{1}{2} f(b) + \sum_{k=1}^{N-1} f(a+kh)]$

```python
import numpy as np

# expect to have I=4.4
def func(x):
    return x**4-2*x+1 

# trapezoidal rule
def int_trapezoidal(func, a, b, N_points):
    dx = (b-a)/N_points
      
    integral = 0.5*func(a) + 0.5*func(b)
    for i in range(1, N_points):
        integral += func(a+i*dx)
    
    integral *= dx
    print(f"Trapezoidal rule: {N_points} points used, I = {integral}")
    return integral

result = int_trapezoidal(func, 0.0, 2.0, 10)
result = int_trapezoidal(func, 0.0, 2.0, 20)
result = int_trapezoidal(func, 0.0, 2.0, 30)
result = int_trapezoidal(func, 0.0, 2.0, 40)
result = int_trapezoidal(func, 0.0, 2.0, 50)
result = int_trapezoidal(func, 0.0, 2.0, 100)
result = int_trapezoidal(func, 0.0, 2.0, 1000)
result = int_trapezoidal(func, 0.0, 2.0, 10000)
```

```python
# trapezoidal rule using numpy array
def int_trapezoidal_NP(func, a, b, N_points):
    dx = (b-a)/N_points
    
    # need to pay attention here about the num of points used for linspace 
    X = np.linspace(a, b, N_points+1)
    Y = func(X)

    integral = Y.sum() - 0.5*func(a) - 0.5*func(b)
    integral *= dx
    print(f"Trapezoidal rule (numpy): {N_points} points used, I = {integral}")
    return integral
```

#Simpson's rule

$I(a, b)=\frac{1}{3}h[f(a) + f(b) + 4\sum_{k=odd, 1, 3, ..., N-1} f(a+kh) + 2\sum_{k=even, 2, 4, ..., N-2} f(a+kh) ]$

```python
import numpy as np

# expect to have I=4.4
def func(x):
    return x**4-2*x+1 

# Simpson's rule, N_points has to be an even number
def int_simpson(func, a, b, N_points):
  integral = 0.

  dx = (b-a)/N_points
  integral = func(a) + func(b)
    
  # for odd terms
  sum_odd_terms = 0.
  for i in range(1, N_points, 2):
    sum_odd_terms += func(a+i*dx)

  # for even terms
  sum_even_terms = 0.
  for i in range(2, N_points, 2):
    sum_even_terms += func(a+i*dx)

  integral += 4*sum_odd_terms;
  integral += 2*sum_even_terms;
  integral = integral*dx/3
  print(f"Simpson rule: {N_points} points used, I = {integral}")

  return integral

result = int_simpson(func, 0.0, 2.0, 10)
result = int_simpson(func, 0.0, 2.0, 20)
result = int_simpson(func, 0.0, 2.0, 30)
result = int_simpson(func, 0.0, 2.0, 40)
result = int_simpson(func, 0.0, 2.0, 50)
result = int_simpson(func, 0.0, 2.0, 100) 
result = int_simpson(func, 0.0, 2.0, 10000)  # already reached the computer precision
result = int_simpson(func, 0.0, 2.0, 100000)  
```

# Excersise 5.4: the diffraction limit of a telescope

Our ability to resolve detail in astronomical observations is limited by the diffraction of light in our telescopes. Light from stars can be treated effectively as coming from a point source at infinity. When such light, with wavelength $\lambda$, passes through the circular aperture of a telescope (which we will assume to have unit radius) and is focused by the telescope in the focal plane, it produces not a single dot, but a circular diffraction pattern consisting of central spot surrounded by a series of concentric rings. The intensity of the light in this diffraction pattern is given by 

$I(r) = \left( \frac{J_1(kr)}{kr}\right)^2$

where $r$ is the distance in the focal plane from the center of the diffraction pattern, $k=2\pi/\lambda$, and $J_1(x)$ is a Bessel function. The Bessel functions $J_m(x)$ are given by

$J_m(x) = \frac{1}{\pi} \int_0^{\pi} \cos (m\theta- x \sin \theta) d\theta$

where $m$ is a nonnegative integer and $x\ge0$.


```python
import numpy as np
import math as math
import matplotlib.pyplot as plt

# define the first Bessel function
def bessel1(r, theta):
    return math.cos(theta - r*math.sin(theta))

# integrate the first Bessel function for a given radius using the trapezoidal rule
def int_bessel1(r):
  N_points = 100
  dtheta = math.pi/N_points
  
  integral = 0.5*bessel1(r, 0) + 0.5*bessel1(r, math.pi)
  for itheta in range(1, N_points):
      integral += bessel1(r, itheta*dtheta)
  integral *= dtheta
  return integral

# make a plot
def plot_J():
    
    # make density plot (500*500 points and we need to do an integral for each point)
    points = 501
    k = 2*math.pi/0.5
    data = np.zeros([points, points])
    x = np.linspace(-1, 1, points)
    y = np.linspace(-1, 1, points)

    for i in range(points):
        for j in range(points):
            r = math.sqrt(x[i]**2 + y[j]**2)
            if abs(r) < 1e-1:
                data[i, j] = 1/4
            else:
                data[i, j] = (int_bessel1(k*r)/(k*r))**2
          
            if(i%100==0 and j%100==0):
              print("Working on point (", i, ", ", j, ")")
            
    plt.xlabel('X ($\mu m$)')
    plt.ylabel('Y ($\mu m$)')
    plt.imshow(data, origin = 'lower', extent = [-1, 1, -1, 1], vmax = 0.01)
    plt.show()

# main program    
plot_J()
```

# Simpson's rule for irregularly spaced data

Suppose the integration interval $I=[a,b]$ needs to be divided into uneven slices (with even number $N$ of slices of widths $h_k$), then the Simpson's rule is given by:
$\int_{a}^{b} f(x) dx = \sum_{i=0}^{N/2-1} \frac{h_{2i}+h_{2i+1}}{6} \left[ (2-\frac{h_{2i+1}}{h_{2i}}) f_{2i} + \frac{(h_{2i}+h_{2i+1})^2}{h_{2i} h_{2i+1}} f_{2i+1} + (2-\frac{h_{2i}}{h_{2i+1}}) f_{2i+2}\right]$

where $f_k$ are the function values at the $k$th sampling point. 

```python
import numpy as np

# function to integrate
def func(x):
  return x**4-2*x+1

# simpson's method on nonuniform points
def simpson_nonuniform(x, f):
    N = len(x) - 1
    h = [x[i+1] - x[i] for i in range(0, N)]
    assert N%2==0

    result = 0.0
    for i in range(1, N, 2):
        h0 = h[i-1]
        h1 = h[i]
        hph = h1 + h0
        hdh = h1/h0
        hmh = h1*h0
        result += (hph/6)*((2-hdh)*f[i-1] + (hph**2/hmh)*f[i] + (2-1/hdh)*f[i+1])

    return result

# bin sizes
X = np.asarray([0., 0.1, 0.3, 0.5, 0.6, 0.7, 0.8, 0.9, 1.1, 1.2, 1.4, 1.8, 2])
Y = func(X)

simpson_nonuniform(X, Y)
```

# Adaptive Trapezoidal rule

$I_i = 0.5 \times I_{i-1} + h_i \sum_{k~odd} f(a+kh_i)$ with $\epsilon_i=(I_i-I_{i-1})/3$

```python
import numpy as np
# expect to have I=4.4
def func(x):
    return x**4-2*x+1 

# calculate the new integrated value when the number of points used increased from N_points to 2*N_points
def int_trapezoidal_adaptive(func, a, b, I_old, N_points):
    dx = (b-a)/(2*N_points)
    
    I_new = 0. 
    for i in range(1, 2*N_points, 2):
        I_new += func(a+i*dx)
    
    I_new *= dx
    I_new += 0.5 * I_old
    
    return I_new

# adaptive trapezoidal rule: once a precision is achieved, it will stop running
def adaptive_trapezoidal(func, a, b, epsilon):
    # define an empty list to store all integrals
    # and initialize the first entry assuming only one bin is used for the integral calculation
    I = []
    I.append(0.5*(func(a)+func(b))*(b-a))
    
    error = 999.
    N_points = 1
    istep = 0
    
    while error > epsilon:
        I.append(int_trapezoidal_adaptive(func, a, b, I[istep], N_points)) 
        error = abs((np.asarray(I[istep+1])-np.asarray(I[istep]))/3)
        N_points *= 2   # double number of points used
        istep += 1      # increase the number of steps by 1
        
    print("All integrated values are: \n", I)
        
adaptive_trapezoidal(func, 0.0, 2.0, 1e-6)
```

# Adaptive Simpson's rule

Define $S_i = \frac{1}{3} [f(a)+f(b) + 2 \sum_{k~even} f(a+kh_i)]$ and $T_i = \frac{2}{3} \sum_{k~odd} f(a+kh_i)$

then we will have $S_i = S_{i-1}+T_{i-1}$ and $I_i=h_i (S_i+2T_i)$

We also have $\epsilon=(I_i-I_{i-1})/15$

```python
import numpy as np

# expect to have I=4.4
def func(x):
  return x**4-2*x+1

# calculate the new integrated value 
# when the number of points used increased from N_points to 2*N_points
def int_simpson_adaptive(func, a, b, S_old, T_old, N_points):
    dx = (b-a)/(2*N_points)
    
    S_new = S_old + T_old    
    T_new = 0.
    for i in range(1, 2*N_points, 2):
        T_new += func(a+i*dx)
    
    T_new *= 2/3.        
    I_new = dx*(S_new + 2*T_new)
    return S_new, T_new, I_new

# adaptive simpson rule: once a precision is achieved, it will stop running
def adaptive_simpson(func, a, b, epsilon):
    # define three empty lists to store S, T and I values
    S, T, I = ([] for i in range(3))   
    
    # initialize the first entries of S, T and I with only two bins used
    S.append(1/3.*(func(a)+func(b)))
    T.append(2/3.*func((a+b)/2.))
    I.append(1/3.*(b-a)/2.*(func(a)+func(b)+4*func((a+b)/2)))
    
    error = 999.
    N_points = 2
    istep = 0
    
    while error > epsilon:
        (S_new, T_new, I_new) = int_simpson_adaptive(func, a, b, S[istep], T[istep], N_points)
        S.append(S_new)
        T.append(T_new)
        I.append(I_new)
        error = abs((I[istep+1]-I[istep])/15.)
        
        N_points *= 2   # double number of points used
        istep += 1      # increase the number of steps by 1
        
    print("All integrated values are: \n", I)
adaptive_simpson(func, 0.0, 2.0, 1e-6)
```

# Excersise 5.4: the diffraction limit of a telescope

Our ability to resolve detail in astronomical observations is limited by the diffraction of light in our telescopes. Light from stars can be treated effectively as coming from a point source at infinity. When such light, with wavelength $\lambda$, passes through the circular aperture of a telescope (which we will assume to have unit radius) and is focused by the telescope in the focal plane, it produces not a single dot, but a circular diffraction pattern consisting of central spot surrounded by a series of concentric rings. The intensity of hte light in this diffraction pattern is given by 

$I(r) = \left( \frac{J_1(kr)}{kr}\right)^2$

where $r$ is the distance in the focal plane from the center of the diffraction pattern, $k=2\pi/\lambda$, and $J_1(x)$ is a Bessel function. The Bessel functions $J_m(x)$ are given by

$J_m(x) = \frac{1}{\pi} \int_0^{\pi} \cos (m\theta- x \sin \theta) d\theta$

where $m$ is a nonnegative integer and $x\ge0$.


```python
import numpy as np
import math as math
import matplotlib.pyplot as plt

# define the Bessel function
def bessel(m, x, theta):
    return math.cos(m*theta - x*math.sin(theta))

# calculate the new integrated value 
# when the number of points used increased from N_points to 2*N_points
def int_simpson_adaptive_bessel(func, m, x, a, b, S_old, T_old, N_points):
    dx = (b-a)/(2*N_points)
    
    S_new = S_old + T_old    
    T_new = 0.
    for i in range(1, 2*N_points, 2):
        T_new += bessel(1, x, i*dx)
    
    T_new *= 2/3.        
    I_new = dx*(S_new + 2*T_new)
    return S_new, T_new, I_new

# adaptive simpson rule  once a precision is achieved, it will stop running
def J_adaptive(m, x, epsilon=1e-6):
    a = 0
    b = math.pi
    
    # define three empty lists to store S, T and I values
    S, T, I = ([] for i in range(3))   
    
    # initialize the first entries of S, T and I with only two bins used
    S.append(1/3.*(bessel(m, x, a)+bessel(m, x, b)))
    T.append(2/3.*bessel(m, x, (a+b)/2.))
    I.append(1/3.*(b-a)/2.*(bessel(m, x, a)+bessel(m, x, b)+4*bessel(m, x, (a+b)/2)))
    
    error = 999.
    N_points = 2
    istep = 0
    
    I_final = 0.
    while error > epsilon:
        (S_new, T_new, I_new) = int_simpson_adaptive_bessel(bessel, m, x, a, b, S[istep], T[istep], N_points)
        S.append(S_new)
        T.append(T_new)
        I.append(I_new)
        error = abs((I[istep+1]-I[istep])/15.)
        I_final = I_new
        
        #print(f"{2*N_points} points used, I = {I[istep+1]}, err = {error}")
        N_points *= 2   # double number of points used
        istep += 1      # increase the number of steps by 1
        
    return I_final

# make a plot
def plot_J():
    
    # make density plot
    points = 501
    k = 2*math.pi/0.5
    data = np.zeros([points, points])
    x = np.linspace(-1, 1, points)
    y = np.linspace(-1, 1, points)

    for i in range(points):
        for j in range(points):
            r = math.sqrt(x[i]**2 + y[j]**2)
            if abs(r) < 1e-1:
                data[i, j] = 1/4
            else:
                data[i, j] = (J_adaptive(1, k*r)/(k*r))**2
            
    plt.xlabel('X ($\mu m$)')
    plt.ylabel('Y ($\mu m$)')
    plt.imshow(data, origin = 'lower', extent = [-1, 1, -1, 1], vmax = 0.01)
    plt.show()

# main program    
plot_J()
```

# Romberg's method

$R_{i,m+1} = R_{i,m} + \frac{1}{4^m-1}(R_{i,m}-R_{i-1,m})$

```python
# expect to have I=4.4
def func(x):
    return x**4-2*x+1 
#    return np.sin(np.sqrt(100*x))**2

# trapezoidal rule
def int_trapezoidal(func, a, b, N_points):
    dx = (b-a)/N_points
      
    integral = 0.5*func(a) + 0.5*func(b)
    for i in range(1, N_points):
        integral += func(a+i*dx)
    
    integral *= dx
    return integral

# Romberg integration
def romberg(func, a, b, nrows):
    
    # define a nrows x nrows array to store the results
    Integ = np.zeros((nrows, nrows))
    for i in range(0, nrows):

        # apply trapezoidal rule for 2^i points
        Integ[i, 0] = int_trapezoidal(func, a, b, 2**i)
      
        # Romberg recursive formula
        for m in range(0, i):
            Integ[i, m+1] = (4**(m+1) * I[i, m] - I[i-1, m]) / (4**(m+1) - 1)

        print(Integ[i, 0:i+1])   # display all results
    return Integ

# main function
p_rows = 4  
Integ = romberg(func, 0, 2, p_rows)
```

**Lagrange polynomials**

```python
import numpy as np
import matplotlib.pyplot as plt

# Lagrange polynomials for a given index k
def lagrange_poly(x, xp, k):
    l = None
    
    # sum over points
    for m in range(len(xp)):
        if m == k:
            continue
        if l == None:
            l = (x - xp[m])/(xp[k] - xp[m])
        else:
            l *= (x - xp[m])/(xp[k] - xp[m])          
    return l

Xp = np.asarray([-2, 1, 4.5, 6.8, 9.5])
Yp = np.asarray([-8.3, -3.9, 1.5, 5.7, 3.4])
plt.plot(Xp, Yp, "ro")

# now make a smooth distribution of the Lagrange polynimials with proper weights applied
X = np.linspace(-2, 10, 101)
Y = []
for iX in range(len(X)):
  Y_value = 0.
  for ip in range(len(Xp)):
    Y_value += lagrange_poly(X[iX], Xp, ip)*Yp[ip] 
  Y.append(Y_value)

plt.plot(X, Y, 'g--')
plt.show()
```

```python
######################################################################
#
# Functions to calculate integration points and weights for Gaussian
# quadrature
#
# x,w = gaussxw(N) returns integration points x and integration
#           weights w such that sum_i w[i]*f(x[i]) is the Nth-order
#           Gaussian approximation to the integral int_{-1}^1 f(x) dx
# x,w = gaussxwab(N,a,b) returns integration points and weights
#           mapped to the interval [a,b], so that sum_i w[i]*f(x[i])
#           is the Nth-order Gaussian approximation to the integral
#           int_a^b f(x) dx
#
# This code finds the zeros of the nth Legendre polynomial using
# Newton's method, starting from the approximation given in Abramowitz
# and Stegun 22.16.6.  The Legendre polynomial itself is evaluated
# using the recurrence relation given in Abramowitz and Stegun
# 22.7.10.  The function has been checked against other sources for
# values of N up to 1000.  It is compatible with version 2 and version
# 3 of Python.
#
# Written by Mark Newman <mejn@umich.edu>, June 4, 2011
# You may use, share, or modify this file freely
#
######################################################################
from numpy import ones,copy,cos,tan,pi,linspace
import math

def gaussxw(N):
    # Initial approximation to roots of the Legendre polynomial
    a = linspace(3,4*N-1,N)/(4*N+2)
    x = cos(pi*a+1/(8*N*N*tan(a)))

    # Find roots using Newton's method
    epsilon = 1e-15
    delta = 1.0
    while delta>epsilon:
        p0 = ones(N,float)
        p1 = copy(x)
        for k in range(1,N):
            p0,p1 = p1,((2*k+1)*x*p1-k*p0)/(k+1)
        dp = (N+1)*(p0-x*p1)/(1-x*x)
        dx = p1/dp
        x -= dx
        delta = max(abs(dx))

    # Calculate the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)

    return x,w

def gaussxwab(N,a,b):
    x,w = gaussxw(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w

def int_gauss_quadrature(func, a, b, N_points):
    # Calculate the sample points and weights, then map them
    # to the required integration domain
    xp, wp = gaussxwab(N_points, a, b)

    # Perform the integration
    integral = 0.0
    for k in range(N_points):
        integral += wp[k]*func(xp[k])

    print(f"Gauss quadrature rule: {N_points} points used, I = {integral}")
    return integral

# integrate x**4-2*x+1 from 0 to 2, the truth value is 4.4
def f(x):
  return x**4-2*x+1 

# integrate x^2 cos(x) from 0 to pi/2, the truth value is 0.4674011
def g(x):
  return x**2*math.cos(x)

print("Integrate x**4-2*x+1, the truth value is 4.4")
val = int_gauss_quadrature(f, 0, 2., 3)
print("\n")

print("Integrate x^2 cos(x), the truth value is 0.4674011")
val = int_gauss_quadrature(g, 0, math.pi/2, 3)
```

**Can also do it in a different way: find zeros for Legendre functions, but integrate the Lagrange polynomials to obtain weights**

```python
# lagrange interpolation example
import math
import numpy as np

# Lagrange polynomials for a given index k
def lagrange_poly(x, xp, k):
       
    l = None
    # sum over points
    for m in range(len(xp)):
        if m == k:
            continue
            
        if l == None:
            l = (x - xp[m])/(xp[k] - xp[m])
        else:
            l *= (x - xp[m])/(xp[k] - xp[m])
                
    return l

# integrate Lagrange polynomials to obtain weights
def int_gauss_quadrature_lagrange_polynomials(func, a, b, N_points):
    
    # Calculate the sample points and weights 
    # weights will not used but we will integrate the Lagrange polynomials to get these weights
    xp, wp = gaussxw(N_points)
    xmap = 0.5*(b-a)*xp + 0.5*(b+a)
   
    # integrate Lagrange polynomials to obtain weights with the Simpson's rule
    wt_lagrange = []
    for k in range(len(xp)):
        N_simpson = 100
        dx = (b-a)/N_simpson
        # we need to use the mapped points for integration here
        integral = lagrange_poly(a, xmap, k) + lagrange_poly(b, xmap, k)
    
        # for odd terms
        sum_odd_terms = 0.
        for i in range(1, N_simpson, 2):
            sum_odd_terms += lagrange_poly(a+i*dx, xmap, k)
    
        # for even terms
        sum_even_terms = 0.
        for i in range(2, N_simpson, 2):
            sum_even_terms += lagrange_poly(a+i*dx, xmap, k)

        integral += 4*sum_odd_terms;
        integral += 2*sum_even_terms;
        integral = integral*dx/3
 
        print("For point ", k, ", wt = ", integral)
        wt_lagrange.append(integral)

    result = 0.
    for k in range(len(xp)):
        result += func(xmap[k])*wt_lagrange[k]

    print(f"Gauss quadrature rule: {N_points} points and Lagrange polynomials used, I = {result}")
    return result

# new function
def func(x):
    return x**2*cos(x)

result1 = int_gauss_quadrature(func, 0, math.pi, 6)
result2 = int_gauss_quadrature_lagrange_polynomials(func, 0., math.pi, 6)
```

#Double integral

$$\int_0^1 dy \int_0^1 dxf(x,y)$$
Define $$F(y) = \int_0^1f(x,y) dx$$
$$\Rightarrow I = \int_0^1 F(y)dy$$


```python
def func(x, y): 
  return pow(pow(x, 4) + pow(y, 5), 0.5) 
  
# double integral
# lx and ux are upper and lower limits of x  
# ly and uy are upper and lower limits of y 
# hx is the step size for integration wrt x 
# hy is the step size for integration wrt y 
def doubleIntegral(func, hx, hy, lx, ux, ly, uy): 
  # numbers of points along the x axis and y axis
  nx = round((ux - lx)/hx + 1) 
  ny = round((uy - ly)/hy + 1)
  
  # z stores the function value at a given (x, y) point
  # ax[] stores the integral wrt y for all x points used
  z = [[0]*ny]*nx   # 2D list
  ax = [0]*nx       # 1D list
  
  # Calculating the values of the function for all grid points
  for i in range(0, nx): 
    for j in range(0, ny): 
      z[i][j] = func(lx+i*hx, ly+j*hy) 
      
  # Calculating the integral value wrt y for each x point
  for i in range(0, nx):
    ax[i] = 0
    for j in range(0, ny):           
      if j == 0 or j == ny - 1: 
        ax[i] += z[i][j] 
      elif j%2 == 0:
        ax[i] += 2 * z[i][j] 
      else:
        ax[i] += 4 * z[i][j] 
    ax[i] *= hy/3

  # now integrate along the x axis  
  answer = 0
  for i in range(0, nx): 
    if i == 0 or i == nx - 1: 
      answer += ax[i] 
    elif i%2 == 0:
      answer += 2 * ax[i] 
    else:
      answer += 4 * ax[i]       
  answer *= hx/3 
  return answer 
  
# lx and ux are upper and lower limits of x  
# ly and uy are upper and lower limits of y 
# hx is the step size for integration wrt x 
# hy is the step size for integration wrt y 
lx, ux, ly, uy = 2.3, 2.5, 3.7, 4.3
hx, hy = 0.01, 0.015
  
answer = doubleIntegral(func, hx, hy, lx, ux, ly, uy)
print(answer)
```

**MC method (easily extendable to multiple integrals)**

$I = \int_a^b \int_c^d f(x, y) dx dy = (b-a)*(d-c)*\frac{1}{N} \sum_{k=1}^{N} f(x_k, y_k)$ where $x_k, y_k$ are randomly-chosen points within the region with $a<x<b$ and $c<y<d$. This method can be extended to higher dimensions.

```python
import random as rn
import numpy as np

# MC method
def integral_MC(f, a, b, c, d, N):
    random_sum = 0.
    
    # loop over random points
    for i in range(N):
        random_x = rn.uniform(a, b)
        random_y = rn.uniform(c, d)
        random_sum += f(random_x, random_y)

    mean = random_sum/N

    return ((b - a)*(d - c)*mean)

# use 1M points for integration
a=2.3
b=2.5
c=3.7
d=4.3
print(integral_MC(func, a, b, c, d, 10**6))
```

**Perform a 10 dimensional integral $\int_0^1 dx_0 \int_0^1 dx_1 ... (x_0+x_1+ ... + x_9)^2$ using the MC method**

```python
import numpy as np

def integral_MC(N):
    integral = 0.
    for i in range(N):
        rand = np.random.random(10)   # generate 10 random numbers between 0 and 1 (since all xs are integrated between 0 and 1)
        median_rand = rand.sum()
        integral += median_rand**2

    # no need to worry about the overall multiplication factor of the ranges 
    # since all xs are integrated between 0 and 1, need to worry about that if the ranges are not 1
    return (integral/N) 

print(integral_MC(1000000))
```

**Integrate a multivariate (n-dimensional) normal function inside a multidimensional sphere**

```python
import scipy.stats as st
import numpy as np

def domain_unit_circle(x):
    # integration domain: sum of x^2 <= 1. 
    # For 2d, it's a unit circle; for 3d it's a unit sphere, etc
    # returns True for inside domain, False for outside
    
    return np.power(x,2).sum() <= 1

def mc_integrate(func, func_domain, a, b, dim, n = 1000):
    # Monte Carlo integration of given function over domain specified by func_domain
    # dim: dimensions of function
    
    # sample x
    x_list = np.random.uniform(a, b, (n, dim))
    
    # determine whether sampled x is inside or outside of domain and calculate its volume
    inside_domain = [func_domain(x) for x in x_list]
    frac_in_domain = sum(inside_domain)/len(inside_domain)
    domain = np.power(b-a, dim) * frac_in_domain
    
    # calculate expected value of func inside domain
    y = func(x_list)
    y_mean = y[inside_domain].sum()/len(y[inside_domain])

    # estimated integration
    integ = domain * y_mean

    return integ

class func_mvn:
    # multivariate normal given mean and covariance
    def __init__(self, mu, cov):
        self.mu = mu
        self.cov = cov
        self.func = st.multivariate_normal(mu, cov)
    
    def calc_pdf(self, x):
        return self.func.pdf(x)

mean = np.array([0, 0.5, 1])
cov = np.array([[1, 0.8, 0.5], [0.8, 1, 0.8], [0.5, 0.8, 1]])

f = func_mvn(mean, cov)
fmvn = f.calc_pdf

print("For multivariate normal, integrated over unit circle")
r = mc_integrate(fmvn, domain_unit_circle, -1, 1, 3, 20000)
print(f"Monte Carlo solution: {r: .4f}")
print(f"Solution from pmvnEll (shotGroups pkg in R): {0.2018934: .4f}")
```
